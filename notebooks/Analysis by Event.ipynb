# Analysis Of Performance

Systematic experiments were run testing various factors affecting performance. The results will be shown comparing the effects of different factors including:

* performance for each individual event
* sentences vs exerpts
* variations of vectorizer
* classifier

## Summary of findings

The results showed that the biggest impact on the results came from:

1) The individual event

2) Sentences vs Excerpts

The differences in the classifiers were minor compared to these effects, though the character based vectorizer, and the linear classifiers were the best overall. The individual event performances vs the performance trained on the full events together was the biggest factor affecting the results. For example, on the sentences, the some individual events could achieve performance of fscore above 0.7, while, in the classifiers trained on all the data together, the maximum performance had fscore around 0.5.
Finally, across both the individual events results and the full datasets together, it was shown that the excerpts performed better than sentence based classifiers by around 0.1 in fscore.

```python
import os
import pandas as pd
```

```python
path = 'classifier_cv_results/'
file_names = os.listdir(path)
file_names
```

## Sentence Based Results

```python
sentence_results_df = pd.DataFrame()

for file_name in file_names:
    if 'sentences' in file_name:
        event_df = pd.read_csv(path+file_name)
        event_df['event'] = file_name.split(' - ')[0]        
        sentence_results_df = pd.concat([sentence_results_df, event_df], ignore_index=True)
```

### Analysis by Event

```python
sentence_results_df.groupby(['event']).fscore.describe()
```

### Analysis by Classifier and Represenation

```python
sentence_results_df.groupby(['classifier', 'vectorizer']).fscore.describe()
```

```python
sentence_results_df.groupby(['classifier']).fscore.describe()
```

```python
sentence_results_df.groupby(['vectorizer']).fscore.describe()
```

## Excerpt Based Results

```python
excerpt_results_df = pd.DataFrame()

for file_name in file_names:
    if 'excerpt_tfidf' in file_name:
        event_df = pd.read_csv(path+file_name)
        event_df['event'] = file_name.split(' - ')[0]        
        excerpt_results_df = pd.concat([excerpt_results_df, event_df], ignore_index=True)
```
