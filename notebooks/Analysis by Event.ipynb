# Analysis Of Performance by Event

```python
import os
import pandas as pd
```

```python
path = 'classifier_cv_results/'
file_names = os.listdir(path)
file_names
```

## Sentence Based Results

```python
sentence_results_df = pd.DataFrame()

for file_name in file_names:
    if 'sentences' in file_name:
        event_df = pd.read_csv(path+file_name)
        event_df['event'] = file_name.split(' - ')[0]        
        sentence_results_df = pd.concat([sentence_results_df, event_df], ignore_index=True)
```

### Analysis by Event

```python
sentence_results_df.groupby(['event']).fscore.describe()
```

### Analysis by Classifier and Represenation

```python
sentence_results_df.groupby(['classifier', 'vectorizer']).fscore.describe()
```

```python
sentence_results_df.groupby(['classifier']).fscore.describe()
```

```python
sentence_results_df.groupby(['vectorizer']).fscore.describe()
```

## Excerpt Based Results

```python
excerpt_results_df = pd.DataFrame()

for file_name in file_names:
    if 'excerpt_tfidf' in file_name:
        event_df = pd.read_csv(path+file_name)
        event_df['event'] = file_name.split(' - ')[0]        
        excerpt_results_df = pd.concat([excerpt_results_df, event_df], ignore_index=True)
```

### Analysis by Event

```python
excerpt_results_df.groupby(['event']).fscore.describe()
```

### Analysis by Classifier and Represenation

```python
excerpt_results_df.groupby(['classifier', 'vectorizer']).fscore.describe()
```

```python
excerpt_results_df.groupby(['classifier']).fscore.describe()
```

```python
excerpt_results_df.groupby(['vectorizer']).fscore.describe()
```
